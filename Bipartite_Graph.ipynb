```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

Teacher = ["Ashiq","Rupok","Sabrina","Mishal"]
Course = ["Capstone","Knowledge","Machine","Cyber","Pattern"]

G.add_nodes_from(Teacher, bipartite=0)
G.add_nodes_from(Course, bipartite=1)

G.add_edge("Ashiq", "Capstone")
G.add_edge("Ashiq", "Knowledge")
G.add_edge("Rupok", "Cyber")
G.add_edge("Rupok", "Pattern")
G.add_edge("Sabrina", "Machine")
G.add_edge("Sabrina", "Knowledge")
G.add_edge("Mishal", "Capstone")
G.add_edge("Mishal", "Pattern")
G.add_edge("Mishal", "Knowledge")
G.add_edge("Sabrina", "Capstone")


nx.draw(G,with_labels=True,node_size=2000)


plt.title("Bipartite Graph: Teacher and Courses")
plt.show()
```

```python
nx.is_bipartite(G)
```

```python
pos = {}
pos.update((n, (0, i)) for i, n in enumerate(Teacher))
pos.update((n, (1, i)) for i, n in enumerate(Course))

nx.draw(G,pos,with_labels=True,node_size=2000)

plt.title("Bipartite Graph: Teacher and Course")
plt.show()
```

```python
from networkx.algorithms import bipartite

set1, set2 = bipartite.sets(G, Teacher)
print("Set 1 (Teacher):", set1)
print("Set 2 (Courses):", set2)
```

# **Knowledge Graph**

```python
from networkx.algorithms import bipartite

student_projection = bipartite.projected_graph(G, Teacher)

nx.draw(student_projection, with_labels=True, node_size=2000)
plt.title("Teacher Projection Graph")
plt.show()
```

```python
ecentricity=nx.eccentricity(G)
print("Ecentricity of each Node: ", ecentricity)
```

```python
diameter=nx.diameter(G)
print("Diameter of the Graph: ",diameter)
```

```python
#radius
radius=nx.radius(G)
print("Radius of the Graph:3 ",radius)
```
